```python
from datasets import load_dataset
import httpx

# Load TechQA dataset
techqa = load_dataset("galileo-ai/ragbench", "techqa", split="test")
print(f"Loaded {len(techqa)} examples")
print(f"Columns: {techqa.column_names}")
```

```python
# Sample a few examples
for i in range(3):
    print(f"\n--- Example {i+1} ---")
    print(f"Question: {techqa[i]['question'][:100]}...")
    print(f"Answer: {techqa[i]['response'][:100]}...")
```

```python
# Query the RAG service
def query_docintel(question: str):
    response = httpx.post(
        "http://localhost:8000/query",
        json={
            "question": question,
            "tenant_id": "demo",
            "user_roles": ["admin"]
        },
        timeout=60.0
    )
    return response.json()

# Test query
result = query_docintel("What is the vacation policy?")
print(result)
```
